```python
def read_expression(filename):
    df = pd.read_table(path+'cases/'+filename, sep='\t', index_col=0, header=[0, 1])
    array = df.to_numpy()
    array = array.T[0]
    return array

# a = read_expression('TARGET-00-NAAENF-11A-01D', 'usc.edu_NBL.HumanMethylation450.Level-3.TARGET-00-NAAENF-11A-01D.txt')
# print(a)
```

```python
l_exp = []
l_cli = []
for sample in samples:
#     file_id = sample[0]
    filename = sample[1]
    case_id = sample[0]
#     print(case_id, filename)
    try:
        death = clinical.loc[case_id].to_numpy()
        expression = read_expression(filename)
#         temp = np.append(expression, death)
#         print(temp)
        l_exp.append(expression)
        l_cli.append(death)
    except:
        print('no such case:', case_id)

print(l_exp)
print(l_cli)
```

```python
exp = np.array(l_exp)
print(exp.shape)
cli = np.array(l_cli)
print(cli.shape)
```

```python
exp_df = pd.DataFrame(exp.T)
print(exp_df.shape)
exp_df.head()
```

```python
m = dict()
c = 1
for i in exp_df.columns:
	if i not in m:
		m[i] = 'F'+str(c)
		c+=1

print(m)
```

```python
exp_df.rename(columns= lambda x:m[x], inplace=True)
exp_df.head()
```

```python
exp_df_copy = exp_df.copy()
exp_df_copy.dropna(how='all', inplace=True)
exp_df_copy.shape
```

```python
exp_df_copy.fillna(0, inplace=True)
```
